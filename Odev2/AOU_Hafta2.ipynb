```python
# 1. İnsan Sınıfını Oluşturun:
class Insan:
    def __init__(self, isim, yas, cinsiyet):
        self.isim = isim
        self._yas = yas  # Kapsülleme için _ ile işaretlendi
        self.cinsiyet = cinsiyet

    def bilgi_ver(self):
        print(f"İsim: {self.isim}, Yaş: {self._yas}, Cinsiyet: {self.cinsiyet}")

    # Yas için getter ve setter metotları
    def get_yas(self):
        return self._yas

    def set_yas(self, yeni_yas):
        if yeni_yas > 0:
            self._yas = yeni_yas
        else:
            print("Yaş negatif olamaz.")

    # 6. Çok Biçimlilik (Polymorphism) için temel konus metodu
    def konus(self):
        pass # Alt sınıflarda yeniden tanımlanacak

# 2. Hoca Sınıfını Oluşturun (Insan'dan Miras Alan):
class Hoca(Insan):
    def __init__(self, isim, yas, cinsiyet, brans):
        super().__init__(isim, yas, cinsiyet)
        self.brans = brans

    def konus(self):
        return f"{self.isim} adlı hoca {self.brans} dersini anlatıyor."

# 3. Öğrenci Sınıfını Oluşturun (Insan'dan Miras Alan):
class Ogrenci(Insan):
    def __init__(self, isim, yas, cinsiyet, sinif, okul_no):
        super().__init__(isim, yas, cinsiyet)
        self.sinif = sinif
        self._okul_no = okul_no # Kapsülleme için _ ile işaretlendi

    def katil(self):
        return f"{self.isim} adlı öğrenci {self.sinif} sınıfında derse katılıyor."

    # Okul_no için getter ve setter metotları
    def get_okul_no(self):
        return self._okul_no

    def set_okul_no(self, yeni_okul_no):
        if yeni_okul_no > 0:
            self._okul_no = yeni_okul_no
        else:
            print("Okul numarası negatif olamaz.")

    # 6. Çok Biçimlilik (Polymorphism) için konus metodu
    def konus(self):
        return f"{self.isim} adlı öğrenci konuşuyor."


# 4. Ana Programı Yazın:
# Nesneleri oluşturma
insan1 = Insan("Ahmet", 30, "Erkek")
hoca1 = Hoca("Ayşe", 45, "Kadın", "Matematik")
ogrenci1 = Ogrenci("Mehmet", 16, "Erkek", "10A", 1234)

# bilgi_ver() metodunu çağırma
print("Nesnelerin Bilgileri:")
insan1.bilgi_ver()
hoca1.bilgi_ver()
ogrenci1.bilgi_ver()

print("\nÖzel Metotları Çağırma:")
# konus() ve katil() metodunu ilgili sınıflar için çağırma
print(hoca1.konus())
print(ogrenci1.katil())
print(ogrenci1.konus()) # Polymorphism örneği

print("\nKapsülleme Örneği (Getter/Setter):")
# Kapsülleme (yas ve okul_no özelliklerini gizleme ve getter/setter kullanma)
print(f"Öğrenci 1'in Okul No (getter ile): {ogrenci1.get_okul_no()}")
ogrenci1.set_okul_no(5678)
print(f"Öğrenci 1'in Yeni Okul No (setter sonrası): {ogrenci1.get_okul_no()}")

print(f"İnsan 1'in Yaşı (getter ile): {insan1.get_yas()}")
insan1.set_yas(35)
print(f"İnsan 1'in Yeni Yaşı (setter sonrası): {insan1.get_yas()}")

# Geçersiz değer atama denemesi
print("\nGeçersiz Değer Atama Denemesi:")
ogrenci1.set_okul_no(-100)
insan1.set_yas(-5)
```

```python
# sorular2 6. soru
from abc import ABC, abstractmethod

# Abstract Özgeçmiş Sınıfı
class AbstractOzgecmis(ABC):
    @abstractmethod
    def ozgecmis_bilgisi(self):
        pass

# Çalışan Ata Sınıfı
class Calisan(AbstractOzgecmis):
    def __init__(self, isim, soyisim, pozisyon):
        self.isim = isim
        self.soyisim = soyisim
        self.pozisyon = pozisyon

    def bilgi_ver(self):
        print(f"İsim: {self.isim}, Soyisim: {self.soyisim}, Pozisyon: {self.pozisyon}")

    # Çalış metodu (metot overriding için temel)
    def calis(self):
        pass # Alt sınıflarda yeniden tanımlanacak

    # Abstract metodu doldurma
    def ozgecmis_bilgisi(self):
        print(f"{self.isim} {self.soyisim} için özgeçmiş bilgisi.")


# Mavi Yaka Çalışan Alt Sınıf
class MaviYaka(Calisan):
    def __init__(self, isim, soyisim, pozisyon, vardiya):
        super().__init__(isim, soyisim, pozisyon)
        self.vardiya = vardiya # Mavi yaka 3 vardiya çalışsın

    # Metot overriding: Mavi yaka için çalış metodu
    def calis(self):
        return f"{self.isim} {self.soyisim} tezgahta/fabrika içinde {self.vardiya}. vardiyada çalışıyor."

    # Abstract metodu doldurma (isteğe bağlı olarak alt sınıfta da özelleştirilebilir)
    def ozgecmis_bilgisi(self):
        print(f"{self.isim} {self.soyisim} (Mavi Yaka) için özgeçmiş bilgisi.")


# Beyaz Yaka Çalışan Alt Sınıf
class BeyazYaka(Calisan):
    def __init__(self, isim, soyisim, pozisyon, vardiya):
        super().__init__(isim, soyisim, pozisyon)
        self.vardiya = vardiya # Beyaz yaka 2 vardiya çalışsın

    # Metot overriding: Beyaz yaka için çalış metodu
    def calis(self):
        return f"{self.isim} {self.soyisim} masabaşında {self.vardiya}. vardiyada çalışıyor."

    # Abstract metodu doldurma (isteğe bağlı olarak alt sınıfta da özelleştirilebilir)
    def ozgecmis_bilgisi(self):
        print(f"{self.isim} {self.soyisim} (Beyaz Yaka) için özgeçmiş bilgisi.")

# Nesneleri türetme (construction ile)
mavi_yaka1 = MaviYaka("Ali", "Veli", "Operatör", 1)
beyaz_yaka1 = BeyazYaka("Ayşe", "Yılmaz", "Mühendis", 2)

# Bilgi verme ve çalışma metotlarını çağırma
print("Çalışan Bilgileri:")
mavi_yaka1.bilgi_ver()
print(mavi_yaka1.calis())
mavi_yaka1.ozgecmis_bilgisi()

print("\nÇalışan Bilgileri:")
beyaz_yaka1.bilgi_ver()
print(beyaz_yaka1.calis())
beyaz_yaka1.ozgecmis_bilgisi()
```
